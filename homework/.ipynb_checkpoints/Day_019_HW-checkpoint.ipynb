```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
#將NA以眾數取代
"""
Your Code Here
"""
from scipy.stats import mode
print("Before replace NAs, numbers of row that HOUSETYPE_MODE is NAs: %i" % sum(app_train['HOUSETYPE_MODE'].isnull()))

value_most = mode(app_train[~app_train['HOUSETYPE_MODE'].isnull()]['HOUSETYPE_MODE'])
print(value_most)

mode_housetype = list(app_train['HOUSETYPE_MODE'].value_counts().index)

app_train.loc[app_train['HOUSETYPE_MODE'].isnull(), 'HOUSETYPE_MODE'] = mode_housetype[0]

print("After replace NAs, numbers of row that HOUSETYPE_MODE is NAs: %i" % sum(app_train['HOUSETYPE_MODE'].isnull()))
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
unique_house_type = app_train['HOUSETYPE_MODE'].unique()

print("All the housetypes are: %s" % unique_house_type)

nrows = len(unique_house_type)
ncols = nrows // 2


plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]), 'TARGET'].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
