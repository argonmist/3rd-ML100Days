```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
def normalize_value(x, y):
    x = 2 * ((x - y.min()) / (y.max() - y.min()) - 0.5) 
    return x
```

```python
"""
Your Code Here
"""
matrix = np.random.random((10,10))
plt.figure(figsize=(10,10))

for i in range(0, 10):
    for j in range(0, 10):
        matrix[i, j] = normalize_value(matrix[i, j], matrix)

sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
plt.title('Heatmap')
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(nrow, ncol)

for i in range(0, nrow):
    for j in range(0, ncol):
        matrix[i, j] = normalize_value(matrix[i, j], matrix)

matrix = pd.DataFrame(matrix)

grid = sns.PairGrid(matrix)
grid = grid.map_upper(plt.scatter, color = 'darkred')
grid = grid.map_lower(sns.kdeplot, cmap = 'Reds')
grid = grid.map_diag(plt.hist, bins = 10, edgecolor =  'k', color = 'darkred');
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(nrow, ncol)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid = grid.map_upper(plt.scatter, color = 'darkred', alpha = 0.2)
grid = grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)
grid = grid.map_diag(plt.hist, bins = 10, edgecolor =  'k', color = 'darkred');
```
