```python
import os
import keras
import itertools
# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
from keras.layers import BatchNormalization
from keras.layers import Dropout

def build_mlp_bn_before(input_shape, output_units = 10, num_neurons = [128, 256, 512], drp_ratio = 0.2, l1_ratio=0.0, l2_ratio=0.0):
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_units in enumerate(num_neurons):
        if i == 0:
            x = BatchNormalization()(input_layer)
            x = keras.layers.Dense(units=n_units, 
                                   activation="relu", 
                                   name="hidden_layer"+str(i+1))(x)
            x = Dropout(drp_ratio)(x)
            
        else:
            x = BatchNormalization()(x)
            x = keras.layers.Dense(units=n_units, 
                                   activation="relu", 
                                   name="hidden_layer"+str(i+1))(x)
            x = Dropout(drp_ratio)(x)
           
    
    out = keras.layers.Dense(units=output_units, activation="softmax", name="output")(x)
    
    model = keras.models.Model(inputs=[input_layer], outputs=[out])
    return model

def build_mlp_bn_after(input_shape, output_units = 10, num_neurons = [128, 256, 512], drp_ratio = 0.2, l1_ratio=0.0, l2_ratio=0.0):
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_units in enumerate(num_neurons):
        if i == 0:        
            x = keras.layers.Dense(units=n_units, 
                                   activation="relu", 
                                   name="hidden_layer"+str(i+1))(input_layer)
            x = Dropout(drp_ratio)(x)
            x = BatchNormalization()(x)
            
        else:
            x = keras.layers.Dense(units=n_units, 
                                   activation="relu", 
                                   name="hidden_layer"+str(i+1))(x)
            x = Dropout(drp_ratio)(x)
            x = BatchNormalization()(x)
            
    
    out = keras.layers.Dense(units=output_units, activation="softmax", name="output")(x)
    
    model = keras.models.Model(inputs=[input_layer], outputs=[out])
    return model
```

```python
LEARNING_RATE = 1e-3
EPOCHS = 20
BATCH_SIZE = 256
MOMENTUM = 0.95
LAYER_NEURONS = [[128, 256, 512], [512, 256, 128]]
Dropout_EXP = [0.0, 0.2]
L1_EXP = [1e-2, 0.0]
L2_EXP = [1e-2, 0.0]
```

```python
def train_mlp(bn_process):
    results = {}

    for i, (layer_neurons, drp_rate, l1r, l2r) in enumerate(itertools.product(LAYER_NEURONS, 
                                                                              Dropout_EXP, 
                                                                              L1_EXP, 
                                                                              L2_EXP)):

        keras.backend.clear_session() # 把舊的 Graph 清掉
        print("Experiment with lay = %s, drp = %s, l1 = %s, l2 = %s" % (str(layer_neurons), 
                                                                        str(drp_rate), 
                                                                        str(l1r), 
                                                                        str(l2r))) 
        if bn_process == "before":
            print("bn before active")
            model = build_mlp_bn_before(input_shape=x_train.shape[1:], 
                                        num_neurons = layer_neurons, 
                                        drp_ratio = drp_rate,
                                        l1_ratio=l1r, 
                                        l2_ratio=l2r)
        if bn_process == "after":
            print("bn after active")
            model = build_mlp_bn_after(input_shape=x_train.shape[1:], 
                                       num_neurons = layer_neurons, 
                                       drp_ratio = drp_rate,
                                       l1_ratio=l1r, 
                                       l2_ratio=l2r)

        model.summary()
        optimizer = keras.optimizers.SGD(lr=LEARNING_RATE, nesterov=True, momentum=MOMENTUM)
        model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer=optimizer)

        model.fit(x_train, y_train, 
                  epochs=EPOCHS, 
                  batch_size=BATCH_SIZE, 
                  validation_data=(x_test, y_test), 
                  shuffle=True)
    
        # Collect results
        train_loss = model.history.history["loss"]
        valid_loss = model.history.history["val_loss"]
        train_acc = model.history.history["accuracy"]
        valid_acc = model.history.history["val_accuracy"]
    
        exp_name_tag = "%s-%s-drp=%s-l1=%s-l2=%s" % (bn_process, 
                                                     str(layer_neurons), 
                                                     str(drp_rate), 
                                                     str(l1r), 
                                                     str(l2r))
        results[exp_name_tag] = {'train-loss': train_loss,
                                 'valid-loss': valid_loss,
                                 'train-acc': train_acc,
                                 'valid-acc': valid_acc}
    return results
```

```python
before_result = train_mlp("before")
```

```python
after_result = train_mlp("after")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def print_result(bn, results):
    color_bar = ["#ff7f50", "#ff4500", "#ff0000", "#008b8b", 
                 "#006400", "#66cdaa", "#00008b", "#ffa500", 
                 "#9932cc", "#808000", "#bdb76b", "#7b68ee",
                 "#20b2aa", "#ffdead", "#8b4513", "#7cfc00"]
    plt.figure(figsize=(24,18))
    for i, cond in enumerate(results.keys()):
        plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond, color=color_bar[i])
        plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond, color=color_bar[i])
    plt.title("%s Loss" % bn)
    plt.legend()
    plt.show()

    plt.figure(figsize=(24,18))
    for i, cond in enumerate(results.keys()):
        plt.plot(range(len(results[cond]['train-acc'])),results[cond]['train-acc'], '-', label=cond, color=color_bar[i])
        plt.plot(range(len(results[cond]['valid-acc'])),results[cond]['valid-acc'], '--', label=cond, color=color_bar[i])
    plt.title("%s Accuracy" % bn)
    plt.legend()
    plt.show()
```

```python
print_result("BN Before Activation", before_result)
```

```python
print_result("BN After Activation", after_result)
```
